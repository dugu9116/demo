```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample sales data based on your input
data = {
    'Telecaller': ['Bhupendra', 'Rahul', 'Gaurav', 'Akash', 'Sadish', 'Pooja'],
    'Total_Sale': [107151, 191885, 10500, 8500, 39840, 50000],
    'Remaining_Target': [5985.50, -91885.50, 89400, 91500, 60160, 100000],
    'Average_Sale': [107151/10, 191885/10, 10500/10, 8500/10, 39840/10, 50000/10],  # Example data
    'Courses_Enrolled': [20, 30, 5, 4, 15, 10],  # Example data
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style
sns.set(style="whitegrid")

```

```python
# 1. Bar plot for Total Sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Telecaller', y='Total_Sale', data=df, palette='viridis')
plt.title('Total Sales by Telecaller for August 2024')
plt.xlabel('Telecaller')
plt.ylabel('Total Sales (with GST)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 2. Bar plot for Remaining Target
plt.figure(figsize=(12, 6))
sns.barplot(x='Telecaller', y='Remaining_Target', data=df, palette='coolwarm')
plt.title('Remaining Target by Telecaller for August 2024')
plt.xlabel('Telecaller')
plt.ylabel('Remaining Target')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# 3. Bar plot for Average Sale
plt.figure(figsize=(12, 6))
sns.barplot(x='Telecaller', y='Average_Sale', data=df, palette='Blues')
plt.title('Average Sale per Course by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Average Sale')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 4. Bar plot for Courses Enrolled
plt.figure(figsize=(12, 6))
sns.barplot(x='Telecaller', y='Courses_Enrolled', data=df, palette='Greens')
plt.title('Number of Courses Enrolled by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Number of Courses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 5. Scatter plot for Total Sales vs. Remaining Target
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Total_Sale', y='Remaining_Target', hue='Telecaller', data=df, s=100)
plt.title('Total Sales vs. Remaining Target')
plt.xlabel('Total Sales (with GST)')
plt.ylabel('Remaining Target')
plt.tight_layout()
plt.show()
```

```python
# 6. Box plot for Total Sales distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='Total_Sale', data=df)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales (with GST)')
plt.tight_layout()
plt.show()
```

```python
# 7. Box plot for Remaining Target distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='Remaining_Target', data=df)
plt.title('Distribution of Remaining Targets')
plt.xlabel('Remaining Target')
plt.tight_layout()
plt.show()
```

```python
# 8. Pie chart for Total Sales distribution by Telecaller
plt.figure(figsize=(8, 8))
plt.pie(df['Total_Sale'], labels=df['Telecaller'], autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Distribution by Telecaller')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# 9. Histogram for Total Sales
plt.figure(figsize=(12, 6))
sns.histplot(df['Total_Sale'], bins=5, kde=True)
plt.title('Histogram of Total Sales')
plt.xlabel('Total Sales (with GST)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 11. Count plot for Courses Enrolled
plt.figure(figsize=(12, 6))
sns.countplot(x='Telecaller', hue='Courses_Enrolled', data=df, palette='Set2')
plt.title('Count of Courses Enrolled by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Count of Courses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 12. Violin plot for Average Sales
plt.figure(figsize=(12, 6))
sns.violinplot(x='Telecaller', y='Average_Sale', data=df, palette='Pastel1')
plt.title('Violin Plot of Average Sales by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Average Sale')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 13. Area plot for Total Sales over Telecallers
plt.figure(figsize=(12, 6))
plt.fill_between(df['Telecaller'], df['Total_Sale'], color="skyblue", alpha=0.4)
plt.plot(df['Telecaller'], df['Total_Sale'], color="Slateblue", alpha=0.6)
plt.title('Total Sales Area Plot by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Total Sales (with GST)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 15. Pairplot for Sales Metrics
sns.pairplot(df)
plt.suptitle('Pairplot of Sales Metrics', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# 16. Bar plot for Total Sales with Error Bars
plt.figure(figsize=(12, 6))
sns.barplot(x='Telecaller', y='Total_Sale', data=df, palette='magma', ci="sd")
plt.title('Total Sales with Error Bars by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Total Sales (with GST)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 17. Line plot for Total Sales over Telecallers
plt.figure(figsize=(12, 6))
sns.lineplot(x='Telecaller', y='Total_Sale', data=df, marker='o', color='blue')
plt.title('Total Sales Trend by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Total Sales (with GST)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 18. FacetGrid for Total Sales by Telecaller
g = sns.FacetGrid(df, col='Telecaller', col_wrap=3, height=4)
g.map(sns.histplot, 'Total_Sale', bins=5)
g.set_titles(col_template="{col_name} Total Sales Histogram")
plt.show()
```

```python

# 19. Stacked Bar Plot for Total Sales and Remaining Target
plt.figure(figsize=(12, 6))
df.set_index('Telecaller')[['Total_Sale', 'Remaining_Target']].plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot for Total Sales and Remaining Target')
plt.xlabel('Telecaller')
plt.ylabel('Sales and Target')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
 #20. Bar plot for Total Sales with percentage labels
plt.figure(figsize=(12, 6))
sns.barplot(x='Telecaller', y='Total_Sale', data=df, palette='viridis')
for index, value in enumerate(df['Total_Sale']):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom')
plt.title('Total Sales by Telecaller with Values')
plt.xlabel('Telecaller')
plt.ylabel('Total Sales (with GST)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 21. Polar Plot for Sales Distribution
plt.figure(figsize=(8, 8), dpi=120)
theta = df['Telecaller']
r = df['Total_Sale']
plt.polar(theta, r, marker='o')
plt.title('Polar Plot of Sales Distribution by Telecaller')
plt.tight_layout()
plt.show()
```

```python
# 22. Cumulative Sales Plot
df_sorted = df.sort_values(by='Total_Sale')
df_sorted['Cumulative_Sales'] = df_sorted['Total_Sale'].cumsum()
plt.figure(figsize=(12, 6))
plt.plot(df_sorted['Telecaller'], df_sorted['Cumulative_Sales'], marker='o')
plt.title('Cumulative Sales by Telecaller')
plt.xlabel('Telecaller')
plt.ylabel('Cumulative Sales (with GST)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
