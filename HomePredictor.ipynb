```python
x=mydata[['price','area','bedrooms','bathrooms','stories','parking']]
```

```python
y=mydata[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
mainroad=LabelEncoder()
guestroom=LabelEncoder()
basement=LabelEncoder()
hotwaterheating=LabelEncoder()
airconditioning=LabelEncoder()
prefarea=LabelEncoder()
furnishingstatus=LabelEncoder()
```

```python
mydata['mainroad']=mainroad.fit_transform(mydata[['mainroad']])
mydata['guestroom']=guestroom.fit_transform(mydata[['guestroom']])
mydata['basement']=basement.fit_transform(mydata[['basement']])
mydata['hotwaterheating']=hotwaterheating.fit_transform(mydata[['hotwaterheating']])
mydata['airconditioning']=airconditioning.fit_transform(mydata[['airconditioning']])
mydata['prefarea']=prefarea.fit_transform(mydata[['prefarea']])
mydata['furnishingstatus']=furnishingstatus.fit_transform(mydata[['furnishingstatus']])
```

```python
mydata
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor  # for regression
# from sklearn.ensemble import RandomForestClassifier  # for classification
```

```python
model = RandomForestRegressor()  # or RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
# For regression
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

```python
print(x_test)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
```
