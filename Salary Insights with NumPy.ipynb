```python
import pandas as pd 
import numpy as np 
```

```python
mydata=pd.read_csv(r'C:\Users\DCS\Desktop\machine learnign data set\Salary Data.csv')
```

```python
mydata
```

```python
# Convert 'Salary' column to a NumPy array
salary_array = np.array(mydata['Salary'])
print(salary_array)
```

```python
# 1. Calculate the mean salary
mean_salary = np.mean(salary_array)
print("Mean Salary:", mean_salary)
```

```python
# 2. Calculate the median salary
median_salary = np.median(salary_array)
print("Median Salary:", median_salary)

```

```python
# 3. Standard deviation of salary
std_salary = np.std(salary_array)
print("Standard Deviation of Salary:", std_salary)
```

```python
# 4. Variance of salary
var_salary = np.var(salary_array)
print("Variance of Salary:", var_salary)
```

```python
# 5. Maximum salary
max_salary = np.max(salary_array)
print("Maximum Salary:", max_salary)
```

```python
# 6. Minimum salary
min_salary = np.min(salary_array)
print("Minimum Salary:", min_salary)
```

```python
# 7. Range of salary (max - min)
range_salary = np.ptp(salary_array)
print("Range of Salary:", range_salary)
```

```python
# 8. Sorting salaries
sorted_salaries = np.sort(salary_array)
print("Sorted Salaries:", sorted_salaries)
```

```python
# 9. Unique salaries
unique_salaries = np.unique(salary_array)
print("Unique Salaries:", unique_salaries)
```

```python
# 10. Reshape the salary array (if applicable)
reshaped_array = salary_array.reshape(-1, 1)
print("Reshaped Array:\n", reshaped_array)
```

```python
# 11. Add a constant value to each salary
salary_plus_bonus = salary_array + 500
print("Salary with Bonus Added:", salary_plus_bonus)
```

```python
# 12. Element-wise multiplication
scaled_salary = salary_array * 1.1
print("Scaled Salary (10% increase):", scaled_salary)
```

```python
# 13. Cumulative sum of salaries
cumsum_salary = np.cumsum(salary_array)
print("Cumulative Sum of Salaries:", cumsum_salary)
```

```python

# 14. Cumulative product of salaries
cumprod_salary = np.cumprod(salary_array)
print("Cumulative Product of Salaries:", cumprod_salary)

```

```python
# 15. Clip salaries to a maximum and minimum value
clipped_salary = np.clip(salary_array, 30000, 100000)
print("Clipped Salary Array:", clipped_salary)

```

```python
# 16. Check for NaN values in salary array
nan_check = np.isnan(salary_array)
print("NaN Check Array:", nan_check)

```

```python
# 17. Replace NaN with the mean salary
salary_array[np.isnan(salary_array)] = mean_salary
print("Salary Array after Replacing NaNs:", salary_array)

```

```python
# 18. Generate random salary data
random_salaries = np.random.randint(30000, 100000, size=10)
print("Randomly Generated Salaries:", random_salaries)

```

```python
# 19. Flatten a multidimensional array
flattened_array = np.ravel(reshaped_array)
print("Flattened Array:", flattened_array)

```

```python
# 20. Create an identity matrix
identity_matrix = np.eye(5)
print("Identity Matrix:\n", identity_matrix)
```
