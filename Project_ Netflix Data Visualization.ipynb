# Data Visualization
Libraries: We'll use libraries like Matplotlib and Seaborn for creating visualizations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")
```

# A. Content Distribution by Type
Visualizing the distribution of content (Movies vs. TV Shows).

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=mydata, x='type', palette='Set2')
plt.title('Content Distribution: Movies vs. TV Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

# Rating Distribution (Histogram)

```python
plt.figure(figsize=(10, 5))
sns.histplot(mydata['rating'], bins=10, kde=True, color='purple')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Content by Country (Bar Chart)

```python
country_counts = mydata['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='Greens_d')
plt.title('Top 10 Countries with Most Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

# Release Trend Over Time (Line Plot)

```python
release_trend = mydata['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x=release_trend.index, y=release_trend.values)
plt.title('Trend of Releases Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.show()
```

# Pie Chart of Content Types

```python
plt.figure(figsize=(8, 8))
mydata['type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Content Types Distribution')
plt.ylabel('')
plt.show()
```

# Ratings vs. Duration (Scatter Plot)

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='duration', y='rating', data=mydata)
plt.title('Ratings vs. Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.show()
```
