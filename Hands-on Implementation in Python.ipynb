```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Step 2: Load a dataset (Iris dataset for this example)
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
print(data)
```

```python
print(X)
```

```python
# Step 3: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
# Step 4: Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X_scaled)
```

```python
print(X_pca)
```

```python
# Step 5: Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Target'] = data.target
```

```python
print(pca_df['Target'])
```

```python
# Step 6: Visualize the PCA results
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']
for target, color in zip(np.unique(data.target), colors):
    plt.scatter(pca_df[pca_df['Target'] == target]['Principal Component 1'],
                pca_df[pca_df['Target'] == target]['Principal Component 2'],
                color=color, alpha=0.5, label=data.target_names[target])
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 7: Explained variance
explained_variance = pca.explained_variance_ratio_
print("Explained variance by component:", explained_variance)
```
