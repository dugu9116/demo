```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=mydata[['order_id','item_price','quantity','transaction_amount']]
```

```python
y=mydata[['date','item_name','item_type','transaction_type','received_by' ,'time_of_sale']]
```

```python
y
```

```python
print(y,mydata)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
date=LabelEncoder()
item_name=LabelEncoder()
item_type=LabelEncoder()
transaction_type=LabelEncoder()
received_by=LabelEncoder()
time_of_sale=LabelEncoder()
```

```python
mydata['date']=date.fit_transform(mydata[['date']])
mydata['item_name']=item_name.fit_transform(mydata[['item_name']])
mydata['item_type']=item_type.fit_transform(mydata[['item_type']])
mydata['transaction_type']=transaction_type.fit_transform(mydata[['transaction_type']])
mydata['received_by']=received_by.fit_transform(mydata[['received_by']])
mydata['time_of_sale']=time_of_sale.fit_transform(mydata[['time_of_sale']])
```

```python
mydata
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
# Train Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.predict([[5,25,8,200]])
```

# # data Visualizations

```python
# Visualizing transaction amounts over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=mydata, x='date', y='transaction_amount', marker='o', color='blue')
plt.title('Transaction Amounts Over Time')
plt.xlabel('Date')
plt.ylabel('Transaction Amount')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Visualizing the relationship between item price and quantity sold
plt.figure(figsize=(12, 6))
sns.scatterplot(data=mydata, x='item_price', y='quantity', hue='item_type', style='transaction_type', palette='deep')
plt.title('Item Price vs Quantity Sold')
plt.xlabel('Item Price')
plt.ylabel('Quantity Sold')
plt.legend()
plt.grid()
plt.show()
```

```python
# Visualizing the distribution of item types
plt.figure(figsize=(12, 6))
sns.countplot(data=mydata, x='item_type', palette='viridis')
plt.title('Distribution of Item Types')
plt.xlabel('Item Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Visualizing the transaction amounts categorized by received by
plt.figure(figsize=(12, 6))
sns.boxplot(data=mydata, x='received_by', y='transaction_amount', palette='Set2')
plt.title('Transaction Amounts by Received By')
plt.xlabel('Received By')
plt.ylabel('Transaction Amount')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Item Price Distribution
plt.figure(figsize=(12, 6))
sns.histplot(mydata['item_price'], bins=30, kde=True, color='green')
plt.title('Item Price Distribution')
plt.xlabel('Item Price')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
 #Customer Purchase Patterns
customer_patterns = mydata.groupby('received_by')['transaction_amount'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=customer_patterns, x='received_by', y='transaction_amount', palette='plasma')
plt.title('Customer Purchase Patterns')
plt.xlabel('Received By')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python

#  Sales Trends for Item Types
plt.figure(figsize=(12, 6))
item_type_trend = mydata.groupby(['date', 'item_type'])['transaction_amount'].sum().unstack()
item_type_trend.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Sales Trends for Item Types')
plt.xlabel('Date')
plt.ylabel('Transaction Amount')
plt.xticks(rotation=45)
plt.grid()
plt.legend(title='Item Type')
plt.show()
```

```python
#Sales Amount vs Item Price
plt.figure(figsize=(12, 6))
sns.scatterplot(data=mydata, x='item_price', y='transaction_amount', hue='item_type', palette='Set1')
plt.title('Sales Amount vs Item Price')
plt.xlabel('Item Price')
plt.ylabel('Transaction Amount')
plt.legend()
plt.grid()
plt.show()
```

```python
#Total Transactions by Date
transactions_per_day = mydata.groupby('date')['order_id'].count().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=transactions_per_day, x='date', y='order_id', color='teal')
plt.title('Total Transactions Per Day')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
 #Quantity Sold by Time of Sale
plt.figure(figsize=(12, 6))
sns.boxplot(data=mydata, x='time_of_sale', y='quantity', palette='muted')
plt.title('Quantity Sold by Time of Sale')
plt.xlabel('Time of Sale')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Most Popular Items Sold
popular_items = mydata['item_name'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=popular_items.index, y=popular_items.values, palette='rocket')
plt.title('Most Popular Items Sold')
plt.xlabel('Item Name')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Transaction Amounts by Transaction Type
plt.figure(figsize=(12, 6))
sns.boxplot(data=mydata, x='transaction_type', y='transaction_amount', palette='Set1')
plt.title('Transaction Amounts by Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Transaction Amount')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
