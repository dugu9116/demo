```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r'C:\Users\DCS\Desktop\machine learnign data set\alary_dataset.csv')
# Display the first few rows of the dataset
print(data.head())
```

```python
print(data.info())
```

```python
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
print(data.describe())
```

```python
# Plot histograms for all numeric columns
data.hist(bins=20, figsize=(15, 10), color='blue', alpha=0.7)
plt.suptitle('Histograms of Numeric Columns')
plt.show()
```

```python
# Box plot for detecting outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=data, orient='h')
plt.title('Box Plots of Numeric Columns')
plt.show()
```

```python
# Scatter plot between 'Experience' and 'Salary' (change as per your columns)
plt.figure(figsize=(10, 6))
plt.scatter(data['YearsExperience'], data['Salary'], alpha=0.5)
plt.title('Scatter Plot of Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.grid()
plt.show()
```

```python
Q1 = data['Salary'].quantile(0.25)  # Change 'Salary' to the relevant column if different
Q3 = data['Salary'].quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data['Salary'] < (Q1 - 1.5 * IQR)) | (data['Salary'] > (Q3 + 1.5 * IQR))]
print("Outliers in the dataset:\n", outliers)
```
