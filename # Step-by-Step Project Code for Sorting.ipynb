```python
import pandas as pd

# Load the dataset
mydata = pd.read_csv(r'C:\Users\IT-Department\Desktop\seattle-weather.csv')
```

```python
mydata
```

```python
# Display the first few rows to understand the structure
print("Original Data:")
print(mydata.head())
```

```python
# Fill missing values if any (you can modify as needed)
mydata = mydata.fillna(method='ffill')
```

```python
print(mydata)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Label encoding the 'weather' column
le = LabelEncoder()
mydata['weather_encoded'] = le.fit_transform(mydata['weather'])
```

```python
print(mydata)
```

```python
# Drop the original 'weather' column
mydata = mydata.drop(columns=['weather'])
```

```python
print("Sorted Data by Date:")
```

```python
# Sort by date for better temporal understanding (optional)
mydata = mydata.sort_values(by='date', ascending=True)
```

```python
print(mydata.head())
```

```python
x= mydata[['precipitation', 'temp_max', 'temp_min', 'wind']]
```

```python
x
```

```python
y= mydata['weather_encoded']
```

```python
y
```

```python
# Selecting features (x) and the target (y)
  # Features
  # Target
```

```python
print(mydata)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into 80% training and 20% testing
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,)
```

```python
print(f"Training Set: {x_train.shape}, Test Set: {x_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model on the training set
rf_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred = rf_model.predict(x_test)
```

```python
# Evaluate the model performance
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Confusion matrix and accuracy
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt 
import numpy as  np 
```

```python
#Get feature importances from the model
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
# Plot the feature importances
plt.figure()
plt.title("Feature Importances")
plt.bar(range(x_train.shape[1]), importances[indices], color="r", align="center")
plt.xticks(range(x_train.shape[1]), x_train.columns[indices], rotation=90)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Example parameter grid for tuning Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
```

```python
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)
```

```python
print(grid_search.best_params_)
# Best parameters from Grid Search
```
