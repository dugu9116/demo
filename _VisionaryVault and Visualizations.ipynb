# Visualization in paroject 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Histogram of a numerical feature (e.g., 'age')


```python
plt.figure(figsize=(10, 6))
sns.histplot(mydata['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 2. Box plot for another numerical feature (e.g., 'charges')


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='charges', data=mydata)
plt.title('Charges by Sex')
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.show()
```

# 3. Scatter plot for two numerical features (e.g., 'age' vs 'charges')

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', data=mydata, hue='smoker', style='sex')
plt.title('Charges vs. Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

# 4. Bar plot for a categorical feature (e.g., 'region')


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='region', data=mydata)
plt.title('Count of Insurance Customers by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

# 9. Pie Chart of Gender Distribution


```python
gender_counts = mydata['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

# 10. Strip Plot of Charges by Age Group

```python
mydata['age_group'] = pd.cut(mydata['age'], bins=[18, 30, 40, 50, 60, 70, 80], labels=['18-30', '31-40', '41-50', '51-60', '61-70', '71-80'])
plt.figure(figsize=(10, 6))
sns.stripplot(x='age_group', y='charges', data=mydata, jitter=True)
plt.title('Charges by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Charges')
plt.show()
```

# 11. Count Plot of Smoker Status by Gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='smoker', data=mydata)
plt.title('Smoker Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# 12. Box Plot of BMI by Smoker Status


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='smoker', y='bmi', data=mydata)
plt.title('BMI by Smoker Status')
plt.xlabel('Smoker')
plt.ylabel('BMI')
plt.show()
```

# 13. Line Plot of Average Charges by Age


```python
avg_charges_age = mydata.groupby('age')['charges'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='charges', data=avg_charges_age)
plt.title('Average Charges by Age')
plt.xlabel('Age')
plt.ylabel('Average Charges')
plt.show()
```

# 14. Bar Plot of Charges by Number of Children


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='children', y='charges', data=mydata, ci=None)
plt.title('Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Charges')
plt.show()
```

# 15. Count Plot of Regions by Smoker Status

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='region', hue='smoker', data=mydata)
plt.title('Smoker Status by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```
