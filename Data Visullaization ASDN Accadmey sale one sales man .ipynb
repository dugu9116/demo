```python
#1. Histogram of Total Course Amount with GST

import matplotlib.pyplot as plt
import seaborn as sns

# Data
total_course_amount = [30000, 30000, 25000, 20000, 11800, 30000, 3000, 50000, 
                       15000, 30000, 30000, 12000, 4000, 12000, 47000, 
                       102.63, 102.63, 102.63, 102.63, 1026.27]

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(total_course_amount, bins=10, kde=True)
plt.title('Distribution of Total Course Amount with GST')
plt.xlabel('Total Course Amount with GST')
plt.ylabel('Frequency')
plt.show()
```

```python
#Histogram of Total Sale (GST amount Included)

total_sale_included = [9097.10, 8000.00, 12500.00, 5000.00, 1000.00, 
                       1000.00, 1500.00, 15097.10, 7597.10, 10000.00, 
                       5000.00, 5000.00, 4000.00, 6097.10, 15000.00, 
                       97.10, 97.10, 97.10, 97.10, 971.00]

plt.figure(figsize=(10, 6))
sns.histplot(total_sale_included, bins=10, kde=True, color='orange')
plt.title('Distribution of Total Sale (GST amount Included)')
plt.xlabel('Total Sale (GST)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar Plot of Total Sales by Course
enroll_courses = ['cyber security specialization', 'cyber security specialization', 
                  'cyber security specialization', 'cyber security specialization', 
                  'ceh batch only', 'cyber security specialization', 
                  'owaps testing guid v 4', 'super 30 batch', 
                  'ceh + vapt only', 'cyber security specialization', 
                  'lets win batch', 'vapt + internship', 
                  'ceh recording course', 'vapt + internship', 
                  'super 30 batch', 'Cyber Security Specialization April Batch', 
                  'Specialization July (Vapt) Batch', 
                  'Academy Batch', 'hacking + kali + linux + parrot security']

# Aggregate total sales by course
course_sales = {}
for course, sale in zip(enroll_courses, total_sale_included):
    if course in course_sales:
        course_sales[course] += sale
    else:
        course_sales[course] = sale

plt.figure(figsize=(12, 6))
sns.barplot(x=list(course_sales.keys()), y=list(course_sales.values()))
plt.title('Total Sales by Course')
plt.xlabel('Enroll Course')
plt.ylabel('Total Sales (GST included)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Pie Chart of Sales Distribution by Course
plt.figure(figsize=(10, 8))
plt.pie(list(course_sales.values()), labels=list(course_sales.keys()), autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Course')
plt.axis('equal')
plt.show()
```

```python
#Box Plot of Total Course Amount with GST
plt.figure(figsize=(10, 6))
sns.boxplot(y=total_course_amount)
plt.title('Box Plot of Total Course Amount with GST')
plt.ylabel('Total Course Amount with GST')
plt.show()
```

# #Line Chart of Total Sales Over Time
Assuming synthetic dates for this example:

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2024-08-01', periods=len(total_sale_included))

plt.figure(figsize=(10, 6))
plt.plot(dates, total_sale_included, marker='o')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales (GST included)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Count Plot of Sales by State
states = ['navi mumbai', 'mathura', 'qatar', 'mendinipur', 'bihar', 
          'rajasthan', 'mauritius', 'maharasthra', 'rajasthan', 
          'japan', 'jharkahnd', 'delhi', 'jharkahand', 
          'uttar pardesh', 'canada', 'canada', 'uttar pradesh', 'gujarat']

plt.figure(figsize=(10, 6))
sns.countplot(y=states)
plt.title('Count of Sales by State')
plt.xlabel('Number of Sales')
plt.ylabel('State')
plt.show()
```

```python
#Heatmap of Total Sales by Telecaller
telecaller_names = ['bhupendra'] * len(total_sale_included)

telecaller_sales = {}
for telecaller, sale in zip(telecaller_names, total_sale_included):
    if telecaller in telecaller_sales:
        telecaller_sales[telecaller] += sale
    else:
        telecaller_sales[telecaller] = sale

plt.figure(figsize=(10, 6))
sns.heatmap([[telecaller_sales['bhupendra']]], annot=True, fmt=".0f", cmap='coolwarm', 
             xticklabels=['Total Sales'], yticklabels=['bhupendra'])
plt.title('Heatmap of Total Sales by Telecaller')
plt.show()
```

```python
#Scatter Plot of Total Course Amount vs. Total Sale
plt.figure(figsize=(10, 6))
plt.scatter(total_course_amount, total_sale_included)
plt.title('Scatter Plot of Total Course Amount vs. Total Sale')
plt.xlabel('Total Course Amount with GST')
plt.ylabel('Total Sale (GST amount Included)')
plt.show()
```

```python
 #Area Chart of Remaining Amounts
remaining_amounts = [0] * len(total_sale_included)  # Assuming no remaining amounts for simplicity

plt.figure(figsize=(10, 6))
plt.fill_between(range(len(remaining_amounts)), remaining_amounts, color='skyblue', alpha=0.5)
plt.title('Area Chart of Remaining Amounts')
plt.xlabel('Sales Entries')
plt.ylabel('Remaining Amount')
plt.show()
```

```python
#Cumulative Sales Over Time
cumulative_sales = np.cumsum(total_course_amount)

plt.figure(figsize=(10, 6))
plt.plot(dates, cumulative_sales, marker='o', color='orange')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales (GST included)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Donut Chart of Total Sales by Course
plt.figure(figsize=(10, 8))
plt.pie(list(course_sales.values()), labels=list(course_sales.keys()), autopct='%1.1f%%', startangle=140, pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart of Sales Distribution by Course')
plt.axis('equal')
plt.show()
```
