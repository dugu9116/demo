```python
import pandas as pd 
```

```python
# Display the first few rows and summary
print(mydata.head())
print(mydata.info())
print(mydata.describe())
```

```python
# Check for missing values
print(mydata.isnull().sum())
```

```python
# Handle missing values (e.g., fill or drop)
mydata.fillna(method='ffill', inplace=True)
print(mydata)
```

```python
# Convert categorical variables to numerical
mydata = pd.get_dummies(mydata, drop_first=True)
print(mydata)
```

```python
# Normalize numerical features if necessary
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
numerical_features = mydata.select_dtypes(include=['float64', 'int64']).columns
mydata[numerical_features] = scaler.fit_transform(mydata[numerical_features])
```

```python
#Determine which features to include in your model.
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(mydata.corr(), annot=True, fmt='.2f')
plt.show()
```

```python
x= mydata.drop('targeted_productivity', axis=1)  # Replace 'productivity' with your actual target column name
```

```python
x
```

```python
print(mydata)
```

```python
y=mydata[['actual_productivity']]
```

```python
y
```

```python
print(mydata)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Train the model
model = RandomForestRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
# Optionally, visualize the predictions
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Productivity')
plt.ylabel('Predicted Productivity')
plt.title('Actual vs Predicted Productivity')
plt.show()
```

```python
#7. Hyperparameter Tuning
#Use Grid Search to optimize the model's parameters.

from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
}
```

```python
grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3)
grid_search.fit(x_train, y_train)
print(f'Best parameters: {grid_search.best_params_}')
```

```python
#Final Model
#Train the final model using the best parameters.
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)
```

```python
# Evaluate the final model
y_final_pred = best_model.predict(x_test)
final_mse = mean_squared_error(y_test, y_final_pred)
print(f'Final Mean Squared Error: {final_mse}')
```
