```python
from sklearn.preprocessing import StandardScaler

# Handle missing values (dropping for simplicity, but you can also impute)
mydata_cleaned = mydata.dropna()

# Select only numeric features for clustering
numeric_features = mydata_cleaned.select_dtypes(include=['float64', 'int64'])

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_features)
```

```python
print(mydata_cleaned)
```

```python
print(numeric_features)
```

```python
print(scaled_data)
```

```python
from sklearn.cluster import KMeans

# Define the number of clusters (k)
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit and predict cluster labels
clusters = kmeans.fit_predict(scaled_data)

# Add the cluster labels back to the original dataframe
mydata_cleaned['Cluster'] = clusters

# Visualize the results
print(mydata_cleaned.head())
```

```python
from sklearn.cluster import DBSCAN

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_clusters = dbscan.fit_predict(scaled_data)

# Add DBSCAN cluster labels to the dataset
mydata_cleaned['DBSCAN_Cluster'] = dbscan_clusters

# Visualize DBSCAN clusters
print(mydata_cleaned.head())
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans, DBSCAN

# K-Means clustering
kmeans = KMeans(n_clusters=3)  # Adjust the number of clusters as needed
kmeans_clusters = kmeans.fit_predict(scaled_data)

# Evaluate K-Means clustering
kmeans_silhouette = silhouette_score(scaled_data, kmeans_clusters)
print(f"Silhouette Score for K-Means: {kmeans_silhouette}")

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps and min_samples as needed
dbscan_clusters = dbscan.fit_predict(scaled_data)

# Evaluate DBSCAN clustering
if len(set(dbscan_clusters)) > 1:
    dbscan_silhouette = silhouette_score(scaled_data, dbscan_clusters)
    print(f"Silhouette Score for DBSCAN: {dbscan_silhouette}")
else:
    print("DBSCAN resulted in a single cluster, silhouette score cannot be calculated.")
```
